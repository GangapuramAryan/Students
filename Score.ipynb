```python
import pandas as pd

df = pd.read_csv("Students.csv")

# APS
df["APS"] = (
    0.5 * (df["gpa"]/10 * 100) +
    0.25 * df["attendance"] +
    0.25 * df["assignments_completion"]
)

# WWS
df["WWS"] = (
    0.4 * df["mental_wellbeing"] * 10 +
    0.3 * df["sleep_hours"] * 10 +
    0.3 * (10 - df["stress_level"]) * 10
)

# PTMS
df["PTMS"] = (
    0.5 * df["productivity_score"] * 10 +
    0.3 * (10 - df["distractions"]) * 10 +
    0.2 * df["attendance"]
)

# CRS
df["CRS"] = (
    0.4 * df["career_clarity"] * 10 +
    0.4 * df["skill_readiness"] * 10 +
    0.2 * df["engagement_score"]
)

# SRI
df["SRI"] = (
    0.25 * df["APS"] +
    0.25 * df["WWS"] +
    0.2 * df["PTMS"] +
    0.2 * df["CRS"] +
    0.1 * df["engagement_score"]
)

# Category
def classify(s):
    if s >= 80:
        return "Green"
    elif s >= 65:
        return "Blue"
    elif s >= 50:
        return "Yellow"
    else:
        return "Red"

df["Category"] = df["SRI"].apply(classify)

df.to_csv("students_scored.csv", index=False)

df.head()
```
